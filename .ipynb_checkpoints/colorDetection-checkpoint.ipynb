## Mini-curso: introdução ao processamento de imagens e visão computacional
### Color detection

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('images/nemo0.jpg')
```

```python
plt.imshow(img)
plt.show()
```

Observe que os canais foram misturados! Eh preciso converter de BGR para RGB.

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

## Vamos plotar a distribuição de cores!

As cores não estão muito bem separadas!

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

```python
from matplotlib.colors import hsv_to_rgb
```

```python
light_color = (21, 255, 255)
dark_color = (1, 100, 100)

lo_square = np.full((10, 10, 3), light_color, dtype=np.uint8) / 255.0
do_square = np.full((10, 10, 3), dark_color, dtype=np.uint8) / 255.0

plt.subplot(1, 2, 1)
plt.imshow(hsv_to_rgb(do_square))
plt.subplot(1, 2, 2)
plt.imshow(hsv_to_rgb(lo_square))
plt.show()
```

```python
mask = cv2.inRange(hsv_img, dark_color, light_color)
result = cv2.bitwise_and(img, img, mask=mask)

plt.subplot(1, 2, 1)
plt.imshow(mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```
